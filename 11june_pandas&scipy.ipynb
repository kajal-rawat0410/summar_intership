```python
#question 13
import pandas as pd
data1= {'emp_id' : [23,45,67,21],
       'name' : ['a','b','c','d']}
data2= {'emp_id': [23,45,67,21],
       'salary': [2345,5600,3400,2300]}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
#result=df1.merge(df2,on='emp_id')
result=pd.merge(df1,df2,on='emp_id',how='inner')
print(result)
```

```python
#sciPy
#question 1
from scipy import optimize
result = optimize.root_scalar(lambda x: x**2 - 4, bracket=[0, 3])
print(result.root)
```
