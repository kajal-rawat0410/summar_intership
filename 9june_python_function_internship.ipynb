```python
#question 1
def fact(n):
  sum=1
  for i in range(1,n+1):
    sum*=i
  return sum

n=int(input("enter number for factorial:"))
print(fact(n))
```

```python
#question 2
def table(n):
  for i in range(1,11):
    print(n,"x",i,"=",n*i)
n=int(input("enter number for multiplication table:"))
print(table(n))
```

```python
#question3
def sumnatural(n):
  sum=0
  for i in range(0,n+1):
    sum+=i
  return sum
n=int(input("enter number of n natural number:"))
print(sumnatural(n))
```

```python
#question4
def isprime(n):
  flag=0
  for i in range(2,n):
    if(n%i==0):
      flag=1
      return True
      break
    else:
      continue
  return False
n=int(input("enter number to check:"))
if(isprime(n)):
  print(n,"is not prime")
else:
  print(n," is prime")
```

```python
#question 5
def largest(lst):
  max=lst[0]
  for i in lst:
    if(i>max):
      max=i
  return max
lst=[]
n=int(input("enter number of users:"))
for i in range(0,n):
  num=int(input("enter element for each user:"))
  lst.append(num)
m=largest(lst)
print("largest number =",m)

```

```python
#question6
def reverse(n):
  sum=0
  for i in range(n,0,-1):
    sum+=i
  return sum
n=int(input("enter number of n natural number:"))
print(reverse(n))
```
